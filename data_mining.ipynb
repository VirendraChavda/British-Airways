```python
import requests
from bs4 import BeautifulSoup
```

## Get the HTML

```python
url = 'https://www.airlinequality.com/airline-reviews/british-airways'
```

```python
req  = requests.get(url)
```

```python
htmlcontent = req.content
```

```python
print(htmlcontent)
```

## Parse the HTML

```python
soup = BeautifulSoup(htmlcontent, 'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
title = soup.title
print(title.get_text())
print(type(title))
```

# Type of object in soup
- Tag
- NavigableString
- BeautifulSoup
- Comment

```python
# get all paragraphs
paras = soup.find_all('p')
print(paras)
```

```python
# get all tags
anchors = soup.find_all('a')
print(anchors)
```

```python
print(soup.find('a'))
```

```python
print(soup.find('a')['rel'])
```

```python
# Get all the links on a page
for link in anchors:
    print(link.get('href'))
```
